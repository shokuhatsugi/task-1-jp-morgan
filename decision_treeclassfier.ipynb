```python
import pandas as pd
df=  pd.read_csv("/content/drive/MyDrive/titanic/titanic.csv")
df.head()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
input= df[['Pclass','Sex', 'Age', 'Fare']]
target= df['Survived']
```

```python
input.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass= LabelEncoder()
le_Sex= LabelEncoder()
le_Age= LabelEncoder()
le_Fare= LabelEncoder()
```

```python
input['Pclass_n']= le_Pclass.fit_transform(input['Pclass'])
input['Sex_n']= le_Sex.fit_transform(input['Sex'])
input['Age_n']= le_Age.fit_transform(input['Age'])
input['Fare_n']= le_Fare.fit_transform(input['Fare'])
input.head()
```

```python
input_n= input.drop(['Pclass'	,'Sex',	'Age',	'Fare'], axis='columns')
```

```python
input_n.head()
```

```python
from sklearn import tree
model= tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

```python
model.score(input_n, target)
```

```python
model.predict([[2,1,28,41]])
```
